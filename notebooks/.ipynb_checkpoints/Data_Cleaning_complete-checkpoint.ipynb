```python
import pandas as pd
```

```python
data = pd.read_csv('../data/raw_data/historical_data_plastic_waste_end_of_life_fates.csv')
#display(data.head(50))
#data.shape

data = data.drop(columns=['Flag Codes', 'Flags', 'TIME'])

display(data.head(50))
data.shape
```

```python
#data = data[data['PLASTICS_EOLFATETYPES'] != 'TOTAL']

#display(data.head(30))
#data.shape
```

```python
print(data.columns)

data['Plastic end of life fates'].unique()
data['PLASTICS_EOLFATETYPES'].unique()

data = data.drop(columns = 'PLASTICS_EOLFATETYPES')
```

```python
data_pivot = data.pivot_table(index=('LOCATION','Location','Time'),columns='Plastic end of life fates', values='Value', aggfunc=sum).reset_index()

data_pivot['Location'].unique()

data_pivot.head()

data_pivot['location_time'] = data_pivot['LOCATION'] + data_pivot['Time'].astype(str)

data_pivot = data_pivot.rename(columns={'Total': 'Total_waste'})

display(data_pivot.head())
data_pivot.shape
```

```python
def drop_not_needed_columns(df: pd.DataFrame) -> pd.DataFrame:

    columns = ['Flag Codes', 'Flags', 'TIME']
    
    df = df.drop(columns = columns, inplace = True)
    return df
```

```python
print(data['LOCATION'].unique())
print(data['Location'].unique())
print(data['LOCATION'].nunique())
print(data['Location'].nunique())
print(data['Time'].nunique())
print(data['Time'].unique())
```

```python
leakage = pd.read_csv('../data/raw_data/Plastic_leakage_from_mismanaged_and_littered_waste.csv')

drop_not_needed_columns(leakage)
leakage['location_time'] = leakage['LOCATION'] + leakage['Time'].astype(str)
leakage = leakage.rename(columns={'Value': 'Total_leakage'})
display(leakage.head())
leakage.shape
```

```python
print(leakage['LOCATION'].unique())
print(leakage['Location'].unique())
print(leakage['LOCATION'].nunique())
print(leakage['Location'].nunique())
print(leakage['Time'].nunique())
print(leakage['Time'].unique())
```

```python
use = pd.read_csv('../data/raw_data/Plastic_use_by_region.csv')

drop_not_needed_columns(use)
use['location_time'] = use['LOCATION'] + use['Time'].astype(str)
use = use.rename(columns={'Value': 'Total_use'})
display(use.head())
use.shape
```

```python
print(use['LOCATION'].unique())
print(use['Location'].unique())
print(use['LOCATION'].nunique())
print(use['Location'].nunique())
print(use['Time'].nunique())
print(use['Time'].unique())
```

```python
waste_rec = pd.read_csv('../data/raw_data/Plastic_waste_collected_for_recycling.csv')

drop_not_needed_columns(waste_rec)

waste_rec['location_time'] = waste_rec['LOCATION'] + waste_rec['Time'].astype(str)
waste_rec = waste_rec.rename(columns={'Value': 'Total_waste_collected_rec'})
display(waste_rec.head())
waste_rec.shape
```

```python
print(waste_rec['LOCATION'].unique())
print(waste_rec['Location'].unique())
print(waste_rec['LOCATION'].nunique())
print(waste_rec['Location'].nunique())
print(waste_rec['Time'].nunique())
print(waste_rec['Time'].unique())
```

```python
plastic_df = pd.merge(data_pivot, leakage, on='location_time')

plastic_df = pd.merge(plastic_df, use, on='location_time')

plastic_df.head()
```

```python
plastic_df = plastic_df.drop(columns=['LOCATION_x', 'Location_x', 'Time_x', 'LOCATION_y', 'Location_y', 'Time_y', 'LOCATION', 'Location', 'Time'])
```

```python
plastic_df = pd.merge(plastic_df, waste_rec, on='location_time')
```

```python
display(plastic_df.head())

print(plastic_df.shape)

print(plastic_df.isna().sum())
```

##### finalizing the data frame
column order and naming convention

```python
# naming convention of column names
plastic_df.columns = plastic_df.columns.str.lower().str.replace(" ","_")
```

```python
# renaming some column names
plastic_df.columns = ['incinerated_waste', 'landfilled_waste', 'littered_waste', 'mismanaged_waste', 'recycled_waste',
       'total_waste', 'location_time', 'total_leakage', 'total_use',
       'location_code', 'location', 'time', 'total_waste_collected_recyc',
       'sum_waste']
print(plastic_df.columns)
```

```python
plastic_df.head(2)
```

```python
#drop again some columns
plastic_df.drop(columns=['sum_waste'], inplace=True)
```

```python
# finally, reorder column oder
plastic_df = plastic_df.reindex(columns=['location_code', 'location', 'time', 'location_time',
                                         'incinerated_waste', 'landfilled_waste', 'littered_waste', 'mismanaged_waste', 'recycled_waste', 'total_waste',
                                         'total_leakage', 'total_use', 'total_waste_collected_recyc'])
```
